```python
#plot scatter and line plots
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import os
import sys
```

```python
embd = ['8','16','32','64','128']
node2vec_auc = [0.535,0.542,0.557,0.570,0.607]
splitter_auc = [0.872,0.898,0.909,0.923,0.931]

node2vec_accuracy= [0.528,0.526,0.540,0.550,0.574]
splitter_accuracy = [0.837,0.855,0.858,0.869,0.877]
```

```python
plt.plot(embd, node2vec_auc,  c = 'r')
plt.plot(embd, splitter_auc, c = 'b' )

#scatter plot

plt.scatter(embd, node2vec_auc, c = 'r')
plt.scatter(embd, splitter_auc, c = 'b' )

plt.xlabel('Embedding Dimension')
plt.ylabel('AUC-ROC')
plt.title('AUC-ROC vs Embedding Dimension for Node2Vec and Splitter')
plt.legend(['Node2Vec','Splitter'])
```

```python
#bar graph with error bars seaborn for dimensions vs accuracy of node2vec and splitter

df = pd.DataFrame({'Embedding Dimension':embd, 'Node2Vec':node2vec_accuracy, 'Splitter':splitter_accuracy})
df = pd.melt(df, id_vars = ['Embedding Dimension'], value_vars = ['Node2Vec','Splitter'])
sns.barplot(x = 'Embedding Dimension', y = 'value', hue = 'variable', data = df, palette = ['r','b'], errorbar='sd',width=0.5, saturation=1)

#y label
plt.ylabel('Accuracy')

#legend
plt.legend()

#title
plt.title('Accuracy vs Embedding Dimension for Node2Vec and Splitter')

```

```python
#bar graph with error bars seaborn for dimensions vs auc-roc of node2vec and splitter

df = pd.DataFrame({'Embedding Dimension':embd, 'Node2Vec':node2vec_auc, 'Splitter':splitter_auc})
df = pd.melt(df, id_vars = ['Embedding Dimension'], value_vars = ['Node2Vec','Splitter'])
sns.barplot(x = 'Embedding Dimension', y = 'value', hue = 'variable', data = df, palette = ['r','b'], errorbar='sd',width=0.5, saturation=1)

#y label
plt.ylabel('AUC-ROC')

#legend
plt.legend()

#title
plt.title('AUC-ROC vs Embedding Dimension for Node2Vec and Splitter')


```
